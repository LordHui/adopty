```python
%matplotlib notebook
import numpy as np
from ipywidgets import interact, FloatSlider
from adopty.utils.viz import get_plot_level_lines
from adopty.datasets import make_coding

_, D, _ = make_coding(n_samples=1, n_atoms=4, n_dim=2)
plot_level_lines = get_plot_level_lines(D)

interact(plot_level_lines, reg=FloatSlider(min=0,max=1,step=.01,value=1),
          theta=FloatSlider(min=-np.pi,max=np.pi,step=.05,value=1.61))

```

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
fig.set_tight_layout(True)
plot_level_lines = get_plot_level_lines(D, fig=fig)

reg = .4
def update(i_frame):
    theta = (2 * i_frame - 1) * np.pi
    plot_level_lines(reg=reg, theta=theta)

# FuncAnimation will call the 'update' function for each frame; here
# animating over 10 frames, with an interval of 200ms between frames.
anim = FuncAnimation(fig, update, frames=np.linspace(0, 1, 100), interval=50)
anim.save('theta.gif', dpi=80, writer='imagemagick')
#     else:
#         # plt.show() will just loop the animation forever.
#         plt.show()
```
